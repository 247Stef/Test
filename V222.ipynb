```python
#Ben√∂tigte Pakete
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.optimize import curve_fit
plt.style.use('seaborn-white')
plt.rcParams['errorbar.capsize']=2
```

```python
freq_mean=np.array([4.49,3.93,3.26,2.72]) #gemittelte Frequenzen
freq_mean_err=np.array([0.003,0.02,0.04,0.01])

eta_th=np.array([0.077,0.071,0.062,0.055]) #thermische Wirkungsgrade
eta_th_err=np.array([0.003,0.003,0.003,0.003])

eta_eff=np.array([0.0172,0.0180,0.0189,0.0202]) #effektive Wirkungsgrade
eta_eff_err=np.array([0.0009,0.0009,0.0009,0.001])
```

```python
def linear_fit(x,a,b):
    return a*x+b
popt1,pcov1=curve_fit(linear_fit,freq_mean,eta_th,sigma=eta_th_err)
popt2,pcov2=curve_fit(linear_fit,freq_mean,eta_eff,sigma=eta_eff_err)
```

```python
plt.plot(freq_mean,eta_th, marker='.',markersize=5, color='black',linewidth=0)
plt.errorbar(freq_mean,eta_th,xerr=freq_mean_err,yerr=eta_th_err,color='black',linewidth=1,linestyle=' ',label=r'Messwerte $\eta_{th}$ mit lin. Fit')
plt.plot(freq_mean,linear_fit(freq_mean,*popt1),color='red',linewidth=0.5)

plt.plot(freq_mean,eta_eff, marker='.',markersize=5, color='blue',linewidth=0)
plt.errorbar(freq_mean,eta_eff,xerr=freq_mean_err,yerr=eta_eff_err,color='blue',linewidth=1,linestyle=' ',label=r'Messwerte $\eta_{eff}$ mit lin. Fit')
plt.plot(freq_mean,linear_fit(freq_mean,*popt2),color='red',linewidth=0.5)

plt.xlabel('Frequenz des Motors f'  ' ' + r' ['+r'$\frac{1}{min}$]', size=15)
plt.ylabel('Wirkungsgrad' + ' ' +'$\eta$', size=15)
plt.title('Diagramm 8: Vergleich von'+ ' ' + r'$\eta_{th}$'+ ' ' + '&' + ' ' + r'$\eta_{eff}$', size=15)
plt.grid(ls='dotted')
plt.legend(frameon=True)
plt.show()

#plt.tight_layout()
#plt.savefig('Desktop/V222Diagramm2.pdf', format='PDF')
```
