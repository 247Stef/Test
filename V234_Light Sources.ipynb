## 3.3 Zuordnung der gefundenen Linien zu Serien

### 3.3.1 Erwartete Linien fr die 1. Nebenserie : md  3p

```python
E_Ry = -13.605
E_3p = -3.026 #[eV]

for m in range(3, 13):
    _3p = 1.2398e3/(E_Ry/m**2 - E_3p)
    
    print('m = {}, _m = {:.1f}' .format(m, _3p))
```

### 3.3.2 Erwartete Linien fr die 2. Nebenserie : ms  3p

```python
E_3s = E_3p - (1.2398e-6/(589e-9)) #[eV]
_s = 3 - np.sqrt(-13.605/E_3s) #correction factor

for m in range(4, 10):
    _3s = 1.2398e3/((E_Ry/(m - _s)**2) - E_3p)
    
    print('m = {}, _m = {:.1f}' .format(m, _3s))
```

### 3.3.3 Erwartete Linien fr die Hauptserie: mp  3s

```python
_p = 3 - np.sqrt(E_Ry/E_3p)

for m in range(4, 6):
    _m = 1.2398e3/((E_Ry/(m - _p)**2) - E_3s)
    
    print('m = {}, _m = {:.1f}' .format(m, _m))
```

## 3.4 Bestimmung der Serienenergien und der l-unabhngigen Korrekturfaktoren

### 1. Nebenserie

```python
#Data input

 = np.array([818.6, 567.9, 497.7, 465.9, 426.0])
d = 2
qn = qn = np.array([3, 4, 5, 6, 11])


#Fit function

def fit_func(m, E_Ry, E_3p, D_d):
    return 1.2398E3/(E_Ry/(m-D_d)**2-E_3p)

par = [-13.6, -3, -0.02]
popt, pcov = curve_fit(fit_func, qn, , sigma = d, p0 = par)
y = np.linspace(2, 13)


#Plot

plt.errorbar(qn, , d, fmt = '.')
plt.plot(y, fit_func(y, *popt), label='Fit', color = 'black', lw = '0.4')
plt.xlim (2.5, 13)
plt.ylim(400, 850)
plt.xlabel('Quantum number')
plt.ylabel('Wavelength [nm]')
plt.title('Diagr. 11: 1. Nebenserie der Natriumatoms')
plt.legend()

fig = plt.gcf()
fig.set_size_inches(11.69, 8.27)
plt.savefig('V234_Fig11.jpg', format = 'jpg', dpi = 600)

print("E_Ry = ({:.3f} +/- {:.3f})eV " .format(popt[0],np.sqrt(pcov[0,0])))
print("E_3p = ({:.3f} +/- {:.3f})eV " .format(popt[1],np.sqrt(pcov[1,1])))
print("D_d = ({:.3f} +/- {:.3f})" .format(popt[2],np.sqrt(pcov[2,2])))
print()


#Quality of fit

chi2_ = np.sum((fit_func(qn,*popt) - )**2/d**2)
dof = len(qn) - 3 #dof:degrees of freedom, Freiheitsgrad
chi2_red = chi2_/dof
print("chi2 =", chi2_)
print("chi2_red =",chi2_red)

#Fit probability

from scipy.stats import chi2
prob = round(1-chi2.cdf(chi2_,dof),2)*100
print("Probability:", prob,"%")
```
